# Task 1 Date Feature Engineering

Create new date-related columns from the transaction_date column.

```python
import pandas as pd
df = pd.read_csv('../data/q3_retail_promotions.csv')

df.head()
```

checked the first few rows.

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

converted the transaction_date column into date format.

This is important because we need to extract year, month, and day information from it.

```python
df['year'] = df['transaction_date'].dt.year
df['month'] = df['transaction_date'].dt.month
df['day_of_week'] = df['transaction_date'].dt.dayofweek
```

The day_of_week column gives the day as a number:

- 0 = Monday
- 1 = Tuesday
- 2 = Wednesday
- 3 = Thursday
- 4 = Friday
- 5 = Saturday
- 6 = Sunday

```python
df['is_month_end'] = 0
df.loc[df['transaction_date'].dt.day >= 25, 'is_month_end'] = 1
```

created a new column called is_month_end.

If the day of the month is 25 or more, I set it to 1.
Otherwise, I set it to 0.

```python

df[['transaction_date', 'year', 'month', 'day_of_week', 'is_month_end']].head()
```

The output shows the new columns created from the transaction_date column.

- year = year
- month = month
- day_of_week = day number
- is_month_end = 0 because date is before 25

This means the feature engineering is correct.
